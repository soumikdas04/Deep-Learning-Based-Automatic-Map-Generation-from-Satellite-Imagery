```python
from PIL import Image
import torch
import cv2
import numpy as np
from diffusers import StableDiffusionXLImg2ImgPipeline
from diffusers import StableDiffusionXLPipeline
```

```python
raw_image_path = f"/kaggle/input/stable-diff-xl-base/epoch_100.png"  # Upload your raw GAN map here
init_image = Image.open(raw_image_path).convert("RGB").resize((1024, 1024))
#init_image.save("input_1280.png")
```

```python
pipe = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0",
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True
).to("cuda")

prompt = "A minimal, flat, clean Google Maps-style map with straight roads, sharp geometry, smooth lines, vector cartography style, high resolution, no distortions, clearly defined structures and borders"
negative_prompt = "blurry, distorted, wavy lines, chaotic layout, sketchy, low resolution"
```

```python
#base = StableDiffusionXLPipeline.from_pretrained(
#    "stabilityai/stable-diffusion-xl-base-1.0",
#    torch_dtype=torch.float16,
#    variant="fp16",
#    use_safetensors=True
#).to("cuda")

#base_image = base(prompt=prompt, image=init_image, strength=0.45, guidance_scale=7.5).images[0]

refined = pipe(
    prompt=prompt,
    image=init_image,
    strength=0.3,
    guidance_scale=8.0,
    negative_prompt=negative_prompt
).images[0]
```
