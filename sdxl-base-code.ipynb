```python
import torch
from diffusers import StableDiffusionXLImg2ImgPipeline
from diffusers.utils import load_image
from PIL import Image
import requests
from io import BytesIO
```

```python
#prompt = "google map styled map, make the wavy lines proper and make the geomerties perfect, make the image minimal and exactly google map style"
```

```python
pipe = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0", torch_dtype=torch.float16, variant="fp16", use_safetensors=True
)
pipe = pipe.to("cuda")
```

```python
url = "/kaggle/input/gan-map-images-reupload/images/epoch_99.png"

init_image = load_image(url).convert("RGB")
init_image = init_image.resize((1280, 1280))
#prompt = "a minimal, flat, clean Google Maps style map with straight roads, accurate geometry, sharp lines,no distortions, high-resolution, modern cartography, vector style, with clear layout and perfect structure, and also corrent the error structures in the map if the road is not correct correct it and any thing seems incorrect correct it and smooth the lines, add borders around the structures if necessary, sharpen the rounded or bended corners"
prompt = "A minimal, flat, clean Google Maps-style map with straight roads, sharp geometry, smooth lines, vector cartography style, high resolution, no distortions, with clearly defined structures and borders"
negative_prompt = "blurry, distorted, wavy lines, chaotic layout, messy, low resolution, sketch, noisy, watercolor"
image = pipe(
    prompt=prompt,
    image=init_image,
    negative_prompt=negative_prompt,
    strength=0.45,
    guidance_scale=8.0,
    num_inference_steps=40
).images[0]
#image = pipe(prompt, image=init_image).images
```

```python
image.save("/kaggle/working/epoch_99.png")
print("✅ Refined image saved as refined_map_output.png")
```
